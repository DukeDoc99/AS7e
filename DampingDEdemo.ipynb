```python
import sympy as sym
sym.init_printing()
```

```python
def eqnsubs(eqns, vartosub, valtosub):
    subeqn = eqns.copy()
    sublist = list(zip(vartosub, valtosub))
    for idx, eqn in enumerate(subeqn):
        subeqn[idx] = subeqn[idx].subs(sublist)
    return subeqn 
```

```python
sym.var('a b c d e f t')
sym.var('x y', cls=sym.Function)
```

```python
eqn1 = sym.Eq(sym.diff(x(t),t),a*x(t)+b*y(t)+c)
eqn2 = sym.Eq(sym.diff(y(t),t),d*x(t)+e*y(t)+f)
eqns = [eqn1, eqn2]
eqns
```

```python
# General Solution
soln=sym.dsolve(eqns)
soln
```

```python
# Underdamped - no forcing in second equation - works, but ugly
eqnsua = eqnsubs(eqns, (a, b, c, d, e, f), (0, -1, 10, 5, -2, 0))
eqnsua
```

```python
solnua = sym.dsolve(eqnsua)
solnua
```

```python
# Underdamped - forcing in second equation - does not work
eqnsub = eqnsubs(eqns, (a, b, c, d, e, f), (0, -1, 10, 5, -2, 3))
eqnsub
```

```python
solnub = sym.dsolve(eqnsub)
solnub
```

```python
# Overdamped - works great
eqnso = eqnsubs(eqns, (a, b, c, d, e, f), (0, -1, 10, 4, -5, 3))
eqnso
```

```python
solno = sym.dsolve(eqnso)
solno
```
